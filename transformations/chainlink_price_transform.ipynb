```python
# get name of each file in chainlink_prices folder without the .csv
files = [f.split('.')[0] for f in os.listdir('data/chainlink_prices')]

# sort files alphabetically
files.sort()
```

```python
# there appears to be some strings within the price table. This was an issue when trying to divide by a float so every row was checked and if it was a string value, then it was set to 0.0
```

```python
counter = 0

# if data/transformed/chainlink_prices folder doesn't exist, then create it
if not os.path.exists('data/transformed/chainlink_prices'):
    os.makedirs('data/transformed/chainlink_prices')
    
for name in files:
    print(f'Round {counter}')
    # load price data from chainlink_prices folder
    df = pd.read_csv(f'data/chainlink_prices/{name}.csv')
    # rename df columns by removing prices_
    df.columns = df.columns.str.replace('prices_', '')
    # add new date column based off of timestamp conversion for chainlink df
    df['date'] = pd.to_datetime(df['timestamp'], unit='s')

    # if a row in price is a string type, replace it with a float 0
    for index, row in df.iterrows():
        if type(row['price']) == str:
            df.at[index, 'price'] = 0.0

    # divide price column by 1e8
    df['price'] = df['price'] / 1e8
    counter += 1

    # save df to data/chainlink_prices folder
    df.to_csv(f'data/transformed/chainlink_prices/{name}.csv', index=False)
    print(df.head(5))
```
