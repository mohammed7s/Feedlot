### Exploratory Data Analysis

```python
print(f'Check if all txHash and settlement values are the same. If True, then they are the same for each row.')
print(trades_df['txHash'].nunique() == trades_df['settlement'].nunique())

# add up buy and sell USD columns
print(f'\ncolumn: buyAmountUsd.sum() = ${trades_df["buyAmountUsd"].sum():,}')
print(f'column: sellAmountUsd.sum() = ${trades_df["sellAmountUsd"].sum():,}')

# difference between buy and sell amounts from above with commas
print(f'buyAmountUsd.sum() - sellAmountUsd.sum() =  ${trades_df["buyAmountUsd"].sum() - trades_df["sellAmountUsd"].sum():,}')
```
